El ordenamiento es clave, porque para una búsqueda secuencial del número 356,321, y en general de cualquier valor $n$, el tiempo de busqueda es del tipo lineal.  La segunda propuesta es un algorítmo de búsqueda binaria. Este algoritmo es particularmente eficiente cuando se trata de conjuntos de datos ordenados, ya que reduce el espacio de búsqueda a la mitad en cada paso. En general, es relevante mencionar que se debe considerar el tiempo de ordenamiento, ya que el algoritmo de búsqueda binaria solamente funciona para este tipo de datos.  Un pseudocódigo propuesto para el algoritmo de búsqueda binaria es el siguiente:

Ahora bien, voy a intentar llevar esto a la práctica:



````python
# primeramente voy a escribir el algoritmo de búsqueda binaria en una función, para llamarla más tarde.

def búsqueda_binaria(lista, objetivo, límite_inferior=None, límite_superior=None): # definimos el algoritmo de búsqueda binaria
  if límite_inferior is None:
      límite_inferior = 0 # Inicio de la lista
  if límite_superior is None:
      límite_superior = len(lista) - 1 # Final de la lista

  if límite_superior < límite_inferior:
      return -1 # devuelve -1 para indicar que el elemento no se encuentra en el rango especificado.

  punto_medio = (límite_inferior + límite_superior) // 2

  if lista[punto_medio] == objetivo:
      return punto_medio
  elif objetivo < lista[punto_medio]:
      return búsqueda_binaria(lista, objetivo, límite_inferior, punto_medio-1)
  else:
      return búsqueda_binaria(lista, objetivo, punto_medio+1, límite_superior)
````

* ## **b. Describa la complejidad del algoritmo que se implementó.**

La complejidad del algoritmo de búsqueda binaria se describe utilizando la notación de Big O $(O)$. La búsqueda binaria es conocida por tener una complejidad de $(O(\log n))$, donde $(n)$ es el tamaño del conjunto de datos. Esta notación indica que el tiempo de ejecución del algoritmo crece logarítmicamente en relación con el tamaño del conjunto de datos.

Vamos a analizarlo brevemente:

- En cada paso de la búsqueda binaria, el tamaño del conjunto de datos se reduce a la mitad. Después de $(k)$ pasos, el tamaño del conjunto será $(n / 2^k)$.

- La búsqueda continúa hasta que el tamaño del conjunto es 1 o el elemento objetivo es encontrado.

- Podemos expresar el número de pasos $(k)$ necesarios para reducir el tamaño del conjunto de datos a 1 como $(n / 2^k = 1)$.

- Resolviendo para $(k)$, obtenemos $(k = \log_2 n)$.

Por lo tanto, la complejidad de la búsqueda binaria es $O(log_2 n)$, ya que el número de pasos necesarios para encontrar el elemento objetivo es proporcional al logaritmo del tamaño del conjunto de datos.

Esto implica que, a medida que el tamaño del conjunto de datos crece, el tiempo de ejecución del algoritmo de búsqueda binaria crece de manera logarítmica y es mucho más eficiente en comparación con algoritmos de búsqueda lineal como la búsqueda secuencial, especialmente para conjuntos de datos grandes.

````python
# voy a reutilizar la función donde se definió el algoritmo de búsqueda binaria, pero hay que escribir ahora uno de la búsqueda secuencial:


def búsqueda_secuencial(lista, objetivo): # definimos el algoritmo de búsqueda secuencial
  for i in range(len(lista)):
      if lista[i] == objetivo:
          return i
  return -1

  punto_medio = (límite_inferior + límite_superior) // 2

  if lista[punto_medio] == objetivo:
      return punto_medio
  elif objetivo < lista[punto_medio]:
      return búsqueda_binaria(lista, objetivo, límite_inferior, punto_medio-1)
  else:
      return búsqueda_binaria(lista, objetivo, punto_medio+1, límite_superior)
````

## **a. Describa las diferencias.**



- El método recursivo implica llamadas a funciones adicionales en cada paso, lo que crea una conjunto de llamdas, o pila de llamadas. Lo que puede causar un desbordamiento de pila si el número es grande.
- El método recursivo utiliza una estructura de función que se llama a sí misma en cada paso.
- El método iterativo utiliza un bucle, lo que evita la acumulación de llamadas en pila, por lo que generalmente es más eficiente en con respecto al uso de memoria.
- el método iterativo utiliza un bucle para calcular el factorial.

Pero aun que esté optimizado para llamadas recursivas, una recursión profunda puede seguir ocupando más recursos que el método iterativo.

## **1. Introducción (Describir las fases del otorgamiento de crédito).**

El otorgamiento de crédito es un proceso crítico en el ámbito financiero que implica evaluar la solvencia y la capacidad de pago de los solicitantes para determinar la viabilidad de conceder un préstamo. Este proceso se divide en varias fases fundamentales, cada una desempeñando un papel crucial en la toma de decisiones$^{[4]}$. La lógica del negocio es simple: Que el prestamista no pierda dinero. A continuación se describen las fases del otorgamiento de crédito:

 1. **Solicitud y Registro:**
   En esta fase, los solicitantes presentan formalmente su solicitud de crédito, proporcionando información personal y financiera. Esta información puede incluir ingresos, historial crediticio,y otros detalles relevantes.

 2. **Evaluación Preliminar:**
   Se realiza una evaluación preliminar para determinar la idoneidad básica del solicitante para el tipo de préstamo solicitado. Se verifican datos como la estabilidad laboral, ingresos o el historial crediticio.

 3. **Recopilación de Información:**
   Los prestamistas recopilan información más detallada sobre el solicitante, incluidos informes de crédito, referencias laborales y financieras, y cualquier otro documento que pueda influir en la decisión de otorgar el crédito.

 4. **Análisis de Riesgos:**
   Se realiza un análisis exhaustivo de los riesgos asociados con el otorgamiento del crédito. Se evalúan factores como la capacidad de pago del solicitante, la estabilidad financiera y las garantías que deja a disposición el solicitante.

 5. **Toma de Decisiones:**
   Con base en la información recopilada y el análisis de riesgos, se toma una decisión sobre la aprobación o denegación del crédito. Esta decisión puede basarse en modelos estadísticos, modelos de Machine Learning, puntajes crediticios o políticas internas del banco prestamista.

 6. **Notificación al Solicitante:**
   El solicitante recibe la notificación formal de la decisión, junto con detalles sobre los términos y condiciones del préstamo en caso de aprobación. En caso de denegación, se debería proporciona información sobre las razones de tal decisión.

 7. **Desembolso del Crédito:**
   Si la solicitud es aprobada, el solicitante recibe los fondos acordados.

 8. **Seguimiento y Gestión del Crédito:**
   Una vez concedido el crédito, se realiza un seguimiento continuo para garantizar el cumplimiento de los términos del préstamo.

*Nota extra:*

$[*]$La prueba de chi-cuadrado $(\chi^2)$ es una prueba estadística que se utiliza para determinar si hay una asociación significativa entre dos variables categóricas. En otras palabras, mide si la distribución observada de frecuencias en una tabla de contingencia es significativamente diferente de la distribución esperada bajo la hipótesis nula de independencia entre las variables.

En el contexto de la correlación entre dos variables categóricas, la hipótesis nula $(H0)$ asume que no hay asociación entre ellas, es decir, son independientes. La hipótesis alternativa $(H1)$ sugiere que hay una asociación significativa.

La fórmula general para el estadístico de chi-cuadrado es:

$ \chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}} $

donde:
- $(O_{ij})$ es la frecuencia observada en la celda $(ij)$ de la tabla de contingencia.
- $(E_{ij})$ es la frecuencia esperada en la celda $(ij)$ bajo la hipótesis nula de independencia.

La prueba de chi-cuadrado compara la suma ponderada de los cuadrados de las diferencias entre las frecuencias observadas y esperadas, normalizadas por las frecuencias esperadas. Si el valor de chi-cuadrado calculado es significativamente grande (según un umbral determinado por el nivel de significancia), se rechaza la hipótesis nula, indicando que hay una asociación significativa entre las variables.

Esta prueba es comúnmente utilizada en análisis de tablas de contingencia y en la evaluación de la bondad de ajuste de un modelo a datos observados.

````python
#Al final decidí realizar la prueba para chi cuadrada

# Hipótesis (H0): Las dos columnas NO están relacionadas entre sí.


from scipy.stats import chi2_contingency

def FuncionChi(inpData, TargetVariable, Lista): # Escribir una función para encontrar la correlación
    SelectedPredictors=[]

    for predictor in Lista:
        CrossTabResult=pd.crosstab(index=inpData[TargetVariable], columns=inpData[predictor])
        ChiSqResult = chi2_contingency(CrossTabResult)

        # If the ChiSq P-Value is <0.05, that means we reject H0
        if (ChiSqResult[1] < 0.05):
            print(predictor, 'está correlacionado con:', TargetVariable, '| P-Value:', ChiSqResult[1])
            SelectedPredictors.append(predictor)
        else:
            print(predictor, 'NO está correlacionado con', TargetVariable, '| P-Value:', ChiSqResult[1])

    return(SelectedPredictors)
````

* ## **3.6 Medir el rendimiento de los modelos.**

Para medir el rendimiento de los dos modelos, se pueden utilizar varias métricas de evaluación del modelo. Aquí, consideraré métricas adicionales:

**Métricas Adicionales:**

1. **Accuracy (Precisión Global):**
   - Mide la proporción de instancias clasificadas correctamente (verdaderos positivos más verdaderos negativos) en relación con el total de instancias.

2. **Precision (Precisión):**
   - Mide la proporción de instancias clasificadas como positivas que realmente son positivas. Es útil cuando el costo de los falsos positivos es alto.

3. **Recall (Recuperación o Sensibilidad):**
   - Mide la proporción de instancias positivas que fueron correctamente identificadas por el modelo. Es útil cuando el costo de los falsos negativos es alto.

4. **F1-Score:**
   - Combina precisión y recall en una sola métrica, proporcionando un equilibrio entre ambas. Es particularmente útil cuando hay un desequilibrio entre las clases.

5. **Área Bajo la Curva ROC (AUC-ROC):**
   - Mide la capacidad del modelo para distinguir entre las clases, representando la relación entre la tasa de verdaderos positivos y la tasa de falsos positivos en varios umbrales de decisión.

Estas métricas ofrecen diferentes perspectivas sobre el rendimiento del modelo y son comúnmente utilizadas en la evaluación de modelos de clasificación binaria. Cada métrica aborda aspectos específicos de la calidad de las predicciones del modelo por lo que proporcionarán una evaluación más completa del rendimiento de los modelos.

````python
# Construir el modelo. Esta es una arquitectura bastante común.
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy']) # Compilar el modelo

model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2) # definimos algunos parámetros típicos como las epocas

loss, accuracy = model.evaluate(X_test, y_test) # pasamos los datos para evaluar
print(f'\nTest Accuracy: {accuracy:.2f}')

y_pred = (model.predict(X_test) > 0.5).astype("int32") #pedimos las predicciones

#imprimims las métricas
print("\nMétricas del Modelo de Red Neuronal:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print(f"Precisión: {precision_score(y_test, y_pred):.2f}")
print(f"Recall: {recall_score(y_test, y_pred):.2f}")
print(f"F1-Score: {f1_score(y_test, y_pred):.2f}")
print(f"AUC-ROC: {roc_auc_score(y_test, y_pred):.2f}")
````

**Red Neuronal 2:**
En esta segunda red neuronal vamos a intentar mejorar la cuestion del sobre ajuste respecto al modelo anterior. La arquitectura de la red neuronal:

1. Capa densa con 128 neuronas, función de activación ReLU y regularización L2 (0.01).
2. Capa de Dropout con una tasa del 50%.
3. Capa densa con 128 neuronas, función de activación ReLU y regularización L2 (0.01).
4. Capa de Dropout con una tasa del 50%.
5. Capa densa con 64 neuronas, función de activación ReLU y regularización L2 (0.01).
6. Capa de Dropout con una tasa del 50%.
7. Capa densa con 16 neuronas, función de activación ReLU y regularización L2 (0.01).
8. Capa densa final con 1 neurona y función de activación sigmoide.

Compilación del modelo:
- Optimizador: Adam con una tasa de aprendizaje de 0.001.
- Función de pérdida: Binary Crossentropy.
- Métricas de evaluación: Accuracy.

Además, se utiliza Early Stopping para evitar el sobreajuste durante el entrenamiento. Evitar el sobre ajuste es una práctica comun en este tipo de modelos.

````python
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Crear modelo de red neuronal con regularización y dropout
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.5))
model.add(Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=Adam(lr=0.001), loss='binary_crossentropy', metrics=['accuracy'])#compilación

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True) # Añadir early stopping para evitar sobreajuste

model.fit(X_train, y_train, epochs=100, batch_size=64, validation_split=0.2, verbose=2, callbacks=[early_stopping])

y_pred = (model.predict(X_test) > 0.5).astype("int32") #pedimos las predicciones

# Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

# Imprimir métricas
print(f"Accuracy: {accuracy:.2f}")
print(f"Precisión: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"AUC-ROC: {roc_auc:.2f}")
````

En comparación con el modelo anterior, se observa una mejora en algunas métricas y una posible mitigación del sobreajuste. Aquí están las observaciones:
